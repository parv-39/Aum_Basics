```python
def aum_diamond(n): #create a function
    if n % 2 == 0: #checks whether the number is even or odd
        print("row must be specified in an odd number.")
        return
    mid = n//2
    count=1
    for i in range(n): #middle
        if i==mid :
            temp = (n-1)//2
            patt = '---'*temp+'AUM'+'---'*temp
            print(patt)
        elif i<mid: #higher pattern
            temp = (n-count)//2
            patt = '---'*temp+'.|.'*count+'---'*temp
            print(patt)
            count +=2
        else:  #lower pattern
            count -=2
            temp = (n-count)//2
            patt = '---'*temp+'.|.'*count+'---'*temp
            print(patt)

row = int(input("Enter an odd number row: "))
aum_diamond(row)
```

```python
def aum_row(n, row, count):
    mid = n // 2

    if row == n:
        return  # base case: stop when all rows are printed

    if row == mid:
        temp = (n - 1) // 2
        patt = '---' * temp + 'AUM' + '---' * temp
        print(patt)
    elif row < mid:
        temp = (n - count) // 2
        patt = '---' * temp + '.|.' * count + '---' * temp
        print(patt)
        count += 2
    else:
        count -= 2
        temp = (n - count) // 2
        patt = '---' * temp + '.|.' * count + '---' * temp
        print(patt)

    # Recursive call to print next row
    aum_row(n, row + 1, count)


def aum_diamond(n):
    if n % 2 == 0:
        print("Row must be specified in an odd number.")
        return

    aum_row(n, row=0, count=1)

row = int(input("Enter an odd number row: "))
aum_diamond(row)
```
