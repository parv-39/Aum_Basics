```python
import matplotlib.pyplot as plt

x = ["Python","C","c++","Java"] #values of x-axis
y = [85,60,65,78] #values of y-axis
plt.xlabel("Languages") #label for x-axis
plt.ylabel("Popularity") #label for y-axis
plt.title("Popularity of Programming Languages") #used to display the title
plt.bar(x,y,label="Usage") #used to display data(x,y) in barchart
# plt.barh(x,y) #used to display data(x,y) in horizontal barchart
# c = ["g","b","r","y"] #used to give multiple colour to bar chart
# plt.bar(x,y,color = c) #used to display data(x,y) in barchart (display with multiple colurs)
# plt.bar(x,y,width = 0.5) #used to display data(x,y) in barchart (display width of particular bar)
plt.xticks(x,rotation = 20) #used to rotate the values of x-axis
plt.legend()
plt.show() #used to display all the data
```

```python
import matplotlib.pyplot as plt

x = [3,6,9,3,1,6,8,4,2] #values of x-axis
y = [10,4,7,2,1,7,9,5,8]  #values of y-axis
colors = ["r","b","g","y","orange","r","g","y","b"] #used to specify color for each scatter plot data
sizes = [400,200,100,300,500,200,300,200,100]
plt.xlabel("X-axis",fontsize = 20) #label for x-axis
plt.ylabel("Y-axis",fontsize = 20) #label for y-axis
plt.scatter(x,y,c = colors,s = sizes,alpha = 0.3,marker = "*",edgecolor = "black") #used to display data(x,y) in scatter plot (colour and size is also specified),(alpha is used to light/bright the colur the data),(marker is used to style the point),(edgecolor is used to give border)
plt.title("Day",fontsize =20) #used to display the title
plt.show() #used to display all the data
```

```python
import matplotlib.pyplot as plt

x = [3,6,9,3,1,6,8,4,2] #values of x-axis
y = [10,4,7,2,1,7,9,5,8]  #values of y-axis
z = [4,7,9,3,6,8,1,2,5]
colors = [10,30,40,56,20,70,90,100,20] #used to specify color for each scatter plot data
sizes = [400,200,100,300,500,200,300,200,100]
plt.xlabel("X-axis",fontsize = 20) #label for x-axis
plt.ylabel("Y-axis",fontsize = 20) #label for y-axis
plt.scatter(x,y,s = sizes,c = colors,cmap = "viridis",alpha = 0.5) #used to display different colour design in graph
plt.scatter(x,z,s = sizes,c = colors,cmap = "afmhot_r",alpha = 1.0) #used to display different colour design in graph
t = plt.colorbar() #used to display the color bar
t.set_label("Colorbar")
plt.title("Day",fontsize =20) #used to display the title
plt.show() #used to display all the data
```

```python
import matplotlib.pyplot as plt

x = [0,10,20,30,40] #values represented on x-axis
v1 = [10,30,50,70,90] #values
v2 = [60,20,40,80,10] #values
v3 = [10,70,20,60,30] #values
l = ["Backend","Data Science","Machine Learning"] #used as a label
c = ["m","g","y"] #used for colour (if not specified then it uses default colours)
plt.stackplot(x,v1,v2,v3,labels = l,colors = c) #it adds the value of v1,v2,v3 and displays using x-axis values
# plt.stackplot(x,v1,v2,v3,labels = l) #it adds the value of v1,v2,v3 and displays using x-axis values
plt.legend(loc=2) #legend is used to display text in graph (label is used in stackplot[as overview]) (loc = location)
plt.title("Python Usage",size = 20) #used to display title with specific size
plt.grid() #used to display gridlines
plt.show() #used to display all the data
```

```python
import matplotlib.pyplot as plt

x = [2020,2021,2022,2023,2024,2025] #values on x-axis
y = [30,10,40,25,60,53] #values on y-axis

plt.step(x,y,color = "r",marker = "o",mfc="g",label = "Performance") #used to display data(x,y) in stepchart(color, marker, label and markercolour are used)
plt.xticks(x,rotation = 20) #used to rotate the 'x' values
plt.legend() #legend is used to display text in graph (label is used in step plot[as overview])
plt.title("Year & Performance",size = 20) #used to display title with specific size
plt.grid() #used to display gridlines
plt.show() #used to display all the data
```
