```python
def check(comp,user):
  if comp == user:
    return 0

  if(comp == 0 and user == 1):
    return -1
  if(comp == 1 and user == 2):
    return -1
  if(comp == 2 and user == 0):
    return -1

    return 1


import random
comp = random.randint(0,2) #used to generate random integer
user = int(input("0 for snake, 1 for water, 2 for gun\n"))

score = check(comp,user)

print(f"You chose{user}\n")
print(f"Computer chose{comp}")

if(score == 0):
    print("Game Draw")
elif(score == -1):
    print("You Lose")
else:
    print("You Won")
```

OOP :- Object-Oriented Programming (OOP) is a way of organizing code using "objects" that represent real things.

Components of OOP in Python


1) Class: The blueprint or template for creating objects. It defines attributes (variables) and methods (functions).

2) Object: An instance of a class containing unique data and behaviors.

3) Encapsulation: Groups related data and methods; protects internal state using access control techniques. (school bag)

4) Inheritance: Enables new classes (child) to acquire properties and behaviors of existing classes (parent), promoting code reuse.(child)

5) Abstraction: Hides unnecessary details and exposes only relevant features for easier usage and interaction.(Coffee)

6) Polymorphism: Allows different classes to provide their own implementation for methods with the same name, enabling flexibility.(phone)

```python
class Shape:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def area(self):
    return self.x * self.y

class Circle(Shape):
  def __init__(self,radius):
    self.radius = radius
    super().__init__(radius,radius)

  def area(self):
    return 3.14 * super().area()

c1 = Circle(5)
print(c1.area())
```

```python
class Person:
  def __init__(self,name):
    self.name = name

class Employee(Person):
  def show(self):
    print(f"{self.name} is an employee")

class Project:
  def __init__(self,project):
    self.project = project

class Teamlead(Employee,Project): #hydrid inheritance
  def __init__(self,name,project):
    Employee.__init__(self,name)
    Project.__init__(self,project)

  def details(self):
    print(f"{self.name} is working on {self.project}")


qw = Teamlead("Keval","Machine learning")
qw.show()
qw.details()


```

```python
import threading
import time
from concurrent.futures import ThreadPoolExecutor

def func(seconds):
  print(f"Sleeping for {seconds} seconds")
  time.sleep(seconds)


def time_val():
  #normal code
  # func(2)
  # func(4)
  # func(6)

  time1 = time.perf_counter() #start of counter


  # #code using thread
  t1 = threading.Thread(target = func,args = [3])
  t2 = threading.Thread(target = func,args = [4])
  t3 = threading.Thread(target = func,args = [7])

  t1.start() #start the compilation process
  t2.start()
  t3.start()

  t1.join() #it will combine all the time delay
  t2.join()
  t3.join()

  time2 = time.perf_counter() #end counter
  print(f"The difference is {time2} and {time1} is :-",time2 - time1) #difference between the counter1 and 2


def poolingdemo():
  with ThreadPoolExecutor() as executer:
    l = [1,3,5,7]
    result = executer.map(func,l)
    for r in result:
      print(r)

time_val()
poolingdemo()
```

```python
#bubble sort
def sort():
  for i in range(len(nums)):
    for j in range(i):
      if nums[j]>nums[j+1]: #checks if 1st value is greater than 2nd value(if yes then below execution is done)
        temp = nums[j] #storing 1st value(greater value) in temporary variable
        nums[j] = nums[j+1] #swaps the 2nd value to num[j]
        nums[j+1] = temp #stores higher value in temp

nums = [8,5,3,7,6,2,9]
nums.sort()
print(nums)
```

```python
import pandas as pd #importing pandas
from fuzzywuzzy import fuzz #importing fuzzywuzzy(fuzzylogic)


def clean_name(name):
    name = name.lower() #used to convert the name to lowercase
    cleaned = '' #empty container
    for char in name: #checks each element of name
        if char.isalpha() or char == ' ': #checks for alphanumeric or space in the name
            cleaned = cleaned + char #if the value is true then it is added to cleaned
    return cleaned #returns the name

def clean_email(email):
    email = email.lower() #used to convert the email to lowercase
    allowed = 'abcdefghijklmnopqrstuvwxyz0123456789@.' #specifies the allowed characters in email
    cleaned = '' #empty container
    for char in email: #checks each element of email
        if char in allowed: #if the elements are present in allowed(category)
            cleaned = cleaned + char #if the value is true then it is added to cleaned
    return cleaned #returns the name

def remove_duplicate(input_file,output_file):
    n = pd.read_csv(input_file) #used to read the input the csv file

    n['clean_name'] = n['name'].map(clean_name) #applies the clean_name function on each name column
    n['clean_email'] = n['email'].map(clean_email) #applies the clean_email function on each email column

    unique_indices = [] #stores the unqiue values
    for email, group in n.groupby('clean_email'): #groups the data by clean_email having one email and compares the same email(groupby :- it is used to aggregate/filtering data)
        seen_names = [] #stores the processed values(seen values)
        for indx,name in zip(group.index,group['clean_name']): #each rows data is combined(index and name from clean_name)
            is_duplicate = False #checks for the duplicate value(and if found, it drops it)
            for i in seen_names:
                if fuzz.ratio(name, i) > 85: #it is used to compare the two filtered names from seen_name and checks that the fuzz ratio is above 85 or not(if not then it checks with other values in dataset)
                    is_duplicate = True #if the fuzz ratio is above 85 then this condition is satisfied
                    break #and it comes out the loop

            if is_duplicate: #if it finds the duplicate value then it goes to next name does the same process
                continue

            seen_names.append(name) #if the name is unique then it adds to name and other names are compared with it
            unique_indices.append(indx)

    index1 = n.loc[unique_indices] #it is used to select the index which contains only unique values
    # index1 = index1.drop(columns=['clean_name','clean_email']) #we can use this line to remove ('clean_name','clean_email')
    index1.to_csv(output_file,index = False) #it is used to move the data in csv file and removes the index vales
    print(f"Deduplicated data saved to '{output_file}'")  #f-string function to display the output

input_file = "C:\extra\parv_assignment2\Customer_Data__With_Duplicates.csv"
remove_duplicate(input_file,'filtered.csv')
```
